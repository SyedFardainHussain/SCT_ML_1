```python
df=df.drop(['Alley','MasVnrType','FireplaceQu','PoolQC','Fence','MiscFeature'],axis=1)
```

```python
#handling remaining missing values
```

```python
df['LotFrontage']=df['LotFrontage'].fillna(df['LotFrontage'].mean())
df['BsmtQual']=df['BsmtQual'].fillna(df['BsmtQual'].mode()[0])
df['BsmtCond']=df['BsmtCond'].fillna(df['BsmtCond'].mode()[0])
df['BsmtExposure']=df['BsmtExposure'].fillna(df['BsmtExposure'].mode()[0])
df['BsmtFinType1']=df['BsmtFinType1'].fillna(df['BsmtFinType1'].mode()[0])
df['BsmtFinType2']=df['BsmtFinType2'].fillna(df['BsmtFinType2'].mode()[0])
df['Electrical']=df['Electrical'].fillna(df['Electrical'].mode()[0])
df['GarageType']=df['GarageType'].fillna(df['GarageType'].mode()[0])
df['GarageYrBlt']=df['GarageYrBlt'].fillna(df['GarageYrBlt'].median())
df['GarageFinish']=df['GarageFinish'].fillna(df['GarageFinish'].mode()[0])
df['GarageQual']=df['GarageQual'].fillna(df['GarageQual'].mode()[0])
df['GarageCond']=df['GarageCond'].fillna(df['GarageCond'].mode()[0])
```

```python
print(df.isna().sum().to_string())
```

```python
plt.figure(figsize=(60,50))
corr=df.corr(numeric_only=True)
sns.heatmap(corr,cmap="Blues",annot=True,fmt=".2%")
plt.show()
```

```python
target_corr=df.corr(numeric_only=True)['SalePrice'].sort_values(ascending=False)
top_features=target_corr.head(5).index.tolist()+target_corr.tail(5).index.tolist()
plt.figure(figsize=(16,6))
plt.title('Top 10 Feature Correlations with SalePrice', fontsize=16, pad=20)
plt.xlabel('House Features', fontsize=12)
plt.ylabel('House Features', fontsize=12)
sns.heatmap(df[top_features].corr(),annot=True,fmt=".2%")

```

```python
#target variable=SalePrice
plt.subplots(figsize=(16,6))
plt.xlabel('Sale Price (USD)')
plt.ylabel('Frequency (Number of Houses)')
plt.title('Distribution of House Sale Prices')
sns.histplot(df['SalePrice'],bins=10,kde=True)
plt.show()
```

```python
#independent features based on the task requirements
features=['GrLivArea','BsmtFullBath','BsmtHalfBath','FullBath','HalfBath','BedroomAbvGr']
#targetvariable=SalePrice
```

```python
#independent feature=X target/dependent feature=y
X=df[features]
y=df['SalePrice']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
X.shape
```

```python
type(X)
```

```python
y=df['SalePrice']
```

```python
y.shape
```

```python
type(y)
```

```python
y=np.array(y).reshape(-1,1)
```

```python
y.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
print(mse)
```

```python
score=r2_score(y_test,y_pred)
print(score*100)
```

```python
# 1. Create a range of values for your main X feature
x_range = np.linspace(X_test.iloc[:,0].min(), X_test.iloc[:,0].max(), 100)

# 2. Create a "fake" X_test where every other column is the mean
X_fake = pd.DataFrame([X_test.mean()] * 100, columns=X_test.columns)
X_fake.iloc[:, 0] = x_range # Replace the first column with our range

# 3. Predict using this "average" data
y_fake_pred = model.predict(X_fake)

# 4. Plot
plt.xlabel('Square Footage (sqft)')
plt.ylabel('Sale Price ($)')
plt.title('House Price vs. Square Footage (Linear Regression)')
plt.scatter(X_test.iloc[:,0], y_test, color='blue', alpha=0.3)
plt.plot(x_range, y_fake_pred, color='pink', linewidth=3)
```
